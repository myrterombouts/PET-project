```python
# {isotope_name : [[(E, I, A), (mu, mu_err), (sigma, sigma_err), (eff, eff_err)], [...], ...]}
CeBrLa_results = {}
CeBr_results   = {}

peak_counter = {"CeBrLa": {}, "CeBr": {}}
```

```python
def model(x, A, mu, sigma, B, C):
    return A / (sigma*np.sqrt(2*np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2)) + B*x + C
```

```python
def make_fit(chnl, cnts, lo, hi, name, detector):
    mask = (chnl > lo) & (chnl< hi)
    x = chnl[mask]
    y = cnts[mask]
    y_err = np.sqrt(y)
    y_err[y_err == 0] = 1
    
    if name not in peak_counter[detector]:
        peak_counter[detector][name] = 0
    else:
        peak_counter[detector][name] += 1
    peak_index = peak_counter[detector][name]
    
    params = isotopes[name]
    energies = params["energies"]
    intensities = params["intensities"]
    activity = params["activity"]
    time_lst = params["time"]
    time = time_lst[0] if detector == "CeBrLa" else time_lst[1]

    peak_index = peak_counter[detector].get(name)
    E_gamma = energies[peak_index]
    I_gamma = intensities[peak_index]

    initial_guess = [max(y), x[np.argmax(y)], 5, 0, min(y)]
    popt, pcov = curve_fit(model, x, y, sigma = y_err, p0=initial_guess, absolute_sigma=True)
    A, mu, sigma, B, C = popt
    A_err, mu_err, sigma_err, B_err, C_err = np.sqrt(np.diag(pcov))
    
    K = activity * I_gamma / 100 * time
    eff = A * 100 / K
    eff_err = A_err * 100 / K
    
    result_entry = [(E_gamma, I_gamma, activity), (mu, mu_err), (sigma, sigma_err), (eff, eff_err)]
    results_dict = CeBrLa_results if detector == "CeBrLa" else CeBr_results
    if name not in results_dict:
        results_dict[name] = []
    results_dict[name].append(result_entry)

    chi_sq = np.sum(((y - model(x, *popt)) / y_err)**2)
    dof = len(x) - len(popt)
    reduced_chi_sq = chi_sq / dof
    p_value = 1 - chi2.cdf(chi_sq, dof)

    plt.figure(figsize=(6,3))
    plt.errorbar(x, y, yerr=y_err, fmt='o', label='Data', capsize=4, color='black', zorder=1)
    plt.plot(x, model(x, *popt), 'r--', linewidth=2, label="Gaussian + linear fit", zorder=3)

    plt.xlabel("Channel")
    plt.ylabel("Counts")
    plt.title(f"{name} peak {peak_index + 1}, {detector}")
    plt.legend()
    plt.show()

    print(f"Counts = {A:.5g} ± {A_err:.5g}")
    print(f"Mean = {mu:.5g} ± {mu_err:.5g}")
    print(f"Sigma = {sigma:.5g} ± {sigma_err:.5g}")
    print(f"Background linear = {B:.5g} ± {B_err:.5g}")
    print(f"Background constant = {C:.5g} ± {C_err:.5g}")
    print("---------------------------------")
    print("Reduced chi squared = ", reduced_chi_sq)
    print("P-value = ", p_value)
```

```python
def sort_peaks(results_dict):
    combined = []
    colors = {
    "Co60": "blue",
    "Cs137": "magenta",
    "Am241": "purple",
    "Eu152": "orange"
    }
    
    for isotope, entries in results_dict.items():
        for entry in entries:
            energy, intesity, activity = entry[0]
            mu, mu_err = entry[1]
            sigma, sigma_err = entry[2]
            eff, eff_err = entry[3]
            color = colors[isotope]

            combined.append({
                "energy" : energy,
                "intesity" : intesity,
                "activity" : activity,
                "mu": mu,
                "mu_err": mu_err,
                "sigma": sigma,
                "sigma_err": sigma_err,
                "eff": eff,
                "eff_err": eff_err,
                "isotope": isotope,
                "color": color
            })

    combined_sorted = sorted(combined, key=lambda x: x["mu"])

    return combined_sorted
```

```python
CeBrLa_fits = sort_peaks(CeBrLa_results)

energies_CeBrLa     = np.array([p["energy"]     for p in CeBrLa_fits])
intesities_CeBrLa   = np.array([p["intesity"]   for p in CeBrLa_fits])
activities_CeBrLa   = np.array([p["activity"]   for p in CeBrLa_fits])
mu_CeBrLa           = np.array([p["mu"]         for p in CeBrLa_fits])
mu_err_CeBrLa       = np.array([p["mu_err"]     for p in CeBrLa_fits])
sigma_CeBrLa        = np.array([p["sigma"]      for p in CeBrLa_fits])
sigma_err_CeBrLa    = np.array([p["sigma_err"]  for p in CeBrLa_fits])
eff_CeBrLa          = np.array([p["eff"]        for p in CeBrLa_fits])
eff_err_CeBrLa      = np.array([p["eff_err"]    for p in CeBrLa_fits])



CeBr_fits = sort_peaks(CeBr_results)

energies_CeBr       = np.array([p["energy"]     for p in CeBr_fits])
intesities_CeBr     = np.array([p["intesity"]   for p in CeBr_fits])
activities_CeBr     = np.array([p["activity"]   for p in CeBr_fits])
mu_CeBr             = np.array([p["mu"]         for p in CeBr_fits])
mu_err_CeBr         = np.array([p["mu_err"]     for p in CeBr_fits])
sigma_CeBr          = np.array([p["sigma"]      for p in CeBr_fits])
sigma_err_CeBr      = np.array([p["sigma_err"]  for p in CeBr_fits])
eff_CeBr            = np.array([p["eff"]        for p in CeBr_fits])
eff_err_CeBr        = np.array([p["eff_err"]    for p in CeBr_fits])
```

```python
def energy_fit(x, A, B, C):
    return A*x**2 + B*x + C
```

```python
x = energies_CeBrLa
y = mu_CeBrLa
y_err = mu_err_CeBrLa

df = pd.DataFrame({"x": x, "y": y, "y_err": y_err})

# standerdazing
x_mean = df['x'].mean()
x_std  = df['x'].std()
y_mean = df['y'].mean()
y_std  = df['y'].std()

df_std = df.copy()
df_std['x'] = (df['x'] - x_mean) / x_std
df_std['y'] = (df['y'] - y_mean) / y_std
df_std['y_err'] = df['y_err'] / y_std

# fit model: A*x^2 + B*x + Intercept
df_std['weights'] = 1 / df_std['y_err']**2

wlsFit = sm.WLS.from_formula(formula='y ~ x + I(x**2)', data=df_std, weights = df_std['weights']).fit()
summ = wlsFit.summary(xname=['Intercept', 'A', 'B'])

# grid for plot
x_grid_std = np.linspace(df_std['x'].min(), df_std['x'].max(), 1000)
grid_df = pd.DataFrame({'x': x_grid_std})

# model
y_fit_std = wlsFit.predict(grid_df)

# 95% CI band
prediction = wlsFit.get_prediction(grid_df)
ci_std = prediction.summary_frame(alpha=0.05)

# undoing the standardizing
x_grid = x_grid_std * x_std + x_mean
y_fit = y_fit_std * y_std + y_mean
ci = ci_std.copy()
ci[['mean','mean_ci_lower','mean_ci_upper']] = ci_std[['mean','mean_ci_lower','mean_ci_upper']] * y_std + y_mean

b = wlsFit.params.values
T = np.array([[
        y_std,
        -y_std * x_mean / x_std,
        y_std * x_mean**2 / x_std**2],
    [
        0,
        y_std / x_std,
        -2 * y_std * x_mean / x_std**2],
    [
        0,
        0,
        y_std / x_std**2]
     ])

alpha = T @ b
alpha[0] += y_mean 
C_CeBrLa, B_CeBrLa, A_CeBrLa = alpha

Cov_b = wlsFit.cov_params().values
Cov_alpha = T @ Cov_b @ T.T
param_errors = np.sqrt(np.diag(Cov_alpha))
C_err_CeBrLa, B_err_CeBrLa, A_err_CeBrLa = param_errors

# residuals
y_fit_data_std = wlsFit.predict(df_std)
y_fit_data = y_fit_data_std * y_std + y_mean
residuals = y - y_fit_data

residual_ci = ci['mean_ci_upper'] - y_fit

# fit quality
N = len(y)
p = len(wlsFit.params)
dof = N - p
chi_sq = np.sum(wlsFit.resid**2 * df_std['weights'])
chi_sq_red = chi_sq / dof
p_value = 1 - chi2.cdf(chi_sq, dof)

# plot
fig, (ax1, ax2) = plt.subplots(
    2, 1, figsize=(8, 6),
    gridspec_kw={'height_ratios': [3, 1]},
    sharex=True
)

used_labels = set()
for i, p in enumerate(CeBrLa_fits):
    label = p["isotope"] if p["isotope"] not in used_labels else None
    used_labels.add(p["isotope"])
    ax1.errorbar(x[i], y[i], yerr=y_err[i], fmt='o', color=p["color"], ecolor='black', capsize=4, label=label)
    
ax1.plot(x_grid, y_fit, 'r--', label=r'model: $Ax^2 + Bx + C$')
ax1.fill_between(x_grid, ci['mean_ci_lower'], ci['mean_ci_upper'], color='grey', alpha=0.25, label='95% CI')
ax1.set_ylabel("Channel")
ax1.legend()
ax1.set_title("CeBrLa Calibration")
ax1.grid(True, alpha=0.3)

for i, p in enumerate(CeBrLa_fits):
    label = p["isotope"] if p["isotope"] not in used_labels else None
    used_labels.add(p["isotope"])
    ax2.errorbar(x[i], residuals[i], yerr=y_err[i], fmt='o', color=p["color"], ecolor='black', capsize=4, label=label)
ax2.axhline(0, color='r', linewidth=1)
ax2.fill_between(x_grid, -residual_ci, residual_ci, color='grey', alpha=0.25, label='95% CI')
ax2.set_xlabel("Energy [keV]")
ax2.set_ylabel("Channel")
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("ENERGY TO CHANNEL")
print(f"Intercept = {C_CeBrLa:.5g} ± {C_err_CeBrLa:.5g}")
print(f"Linear = {B_CeBrLa:.5g} ± {B_err_CeBrLa:.5g}")
print(f"Quadratic = {A_CeBrLa:.5g} ± {A_err_CeBrLa:.5g}")
print("----------------------")
print(f"Reduced chi-squared = {chi_sq_red:.5g}")
print(f"P value = {p_value:.5g}")
```

```python
def channel_to_energy(channel, channel_err, a, b, c, a_err, b_err, c_err):
    channel = np.asarray(channel)
    
    D = b**2 - 4*a*(c - channel)
    
    if np.any(D < 0):
        raise ValueError("Negative discriminant encountered.")
    
    sqrtD = np.sqrt(D)   
    E = (-b + sqrtD) / (2*a)

    dE_da = -(c - channel)/(a*sqrtD) - E/a
    dE_db = -1/(2*a) + b/(2*a*sqrtD)
    dE_dc = -1/sqrtD
    dE_dch = 1/sqrtD
    
    E_err = np.sqrt(
        (dE_da * a_err)**2 +
        (dE_db * b_err)**2 +
        (dE_dc * c_err)**2 +
        (dE_dch * channel_err)**2
    )
    
    return E, E_err
```

```python
def calc_E_res(mu, mu_err, sigma, sigma_err, a, b, c, a_err, b_err, c_err):
    E, E_err = channel_to_energy(mu, mu_err, a, b, c, a_err, b_err, c_err)
    
    sigma_E, sigma_E_err = channel_to_energy(sigma, sigma_err, a, b, c, a_err, b_err, c_err)
    
    FWHM = 2.355*sigma_E
    FWHM_err = 2.355*sigma_E_err
    
    E_res = FWHM/E * 100
    E_res_err = np.sqrt((FWHM_err / E)**2 + (FWHM * E_err / E**2)**2) * 100
    
    return E_res, E_res_err
```

```python
E_res_CeBrLa, E_res_err_CeBrLa = calc_E_res(mu_CeBrLa, mu_err_CeBrLa, sigma_CeBrLa, sigma_err_CeBrLa, 
                                            A_CeBrLa, B_CeBrLa, C_CeBrLa, A_err_CeBrLa, B_err_CeBrLa, C_err_CeBrLa)

E_res_CeBr, E_res_err_CeBr = calc_E_res(mu_CeBr, mu_err_CeBr, sigma_CeBr, sigma_err_CeBr, 
                                            A_CeBr, B_CeBr, C_CeBr, A_err_CeBr, B_err_CeBr, C_err_CeBr)
```

```python
def make_fit_Na22(chnl, cnts, lo, hi, pos, detector, time):
    mask = (chnl > lo) & (chnl< hi)
    x = chnl[mask]
    y = cnts[mask]
    y_err = np.sqrt(y)
    y_err[y_err == 0] = 1

    initial_guess = [max(y), x[np.argmax(y)], 5, 0, min(y)]
    popt, pcov = curve_fit(model, x, y, sigma = y_err, p0=initial_guess, absolute_sigma=True)
    A, mu, sigma, B, C = popt
    A_err, mu_err_, sigma_err, B_err, C_err = np.sqrt(np.diag(pcov))
    mu_err = np.sqrt(mu_err_**2 + 0.5**2) # systematic error of 0.5

    intensity = 179.8
    activity = 573.9
    K = activity * intensity / 100 * time
    eff = abs(A) * 100 / K
    eff_err = abs(A_err) * 100 / K

    if detector == 'CeBrLa':
        eff_CeBrLa.append([eff, eff_err])
    else:
        eff_CeBr.append([eff, eff_err])

    chi_sq = np.sum(((y - model(x, *popt)) / y_err)**2)
    dof = len(x) - len(popt)
    reduced_chi_sq = chi_sq / dof
    p_value = 1 - chi2.cdf(chi_sq, dof)

    plt.figure(figsize=(6,3))
    plt.errorbar(x, y, yerr=y_err, fmt='o', label='Data', capsize=4, color='black', zorder=1)
    plt.plot(x, model(x, *popt), 'r--', linewidth=2, label="Gaussian + linear fit", zorder=3)

    plt.xlabel("Channel")
    plt.ylabel("Counts")
    plt.title(f"Na22 peak pos {pos+1}, {detector}")
    plt.legend()
    plt.show()

    print(f"Counts = {A:.5g} ± {A_err:.5g}")
    print(f"Efficiency = {eff:.5g} ± {eff_err:.5g}")
    print(f"Mean = {mu:.5g} ± {mu_err_:.5g} (± {mu_err:.5g})")
    print(f"Sigma = {sigma:.5g} ± {sigma_err:.5g}")
    print(f"Background linear = {B:.5g} ± {B_err:.5g}")
    print(f"Background constant = {C:.5g} ± {C_err:.5g}")
    print("---------------------------------")
    print("Reduced chi squared = ", reduced_chi_sq)
    print("P-value = ", p_value)
```
