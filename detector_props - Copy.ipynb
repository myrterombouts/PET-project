```python
# {isotope_name : [[(E, I, A), (mu, mu_err), (sigma, sigma_err), (eff, eff_err)], [...], ...]}
CeBrLa_results = {}
CeBr_results   = {}

peak_counter = {"CeBrLa": {}, "CeBr": {}}
```

```python
def model(x, A, mu, sigma, B, C):
    return A / (sigma*np.sqrt(2*np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2)) + B*x + C
```

```python
def make_fit(chnl, cnts, lo, hi, name, detector):
    mask = (chnl > lo) & (chnl< hi)
    x = chnl[mask]
    y = cnts[mask]
    y_err = np.sqrt(y)
    y_err[y_err == 0] = 1
    
    if name not in peak_counter[detector]:
        peak_counter[detector][name] = 0
    else:
        peak_counter[detector][name] += 1
    peak_index = peak_counter[detector][name]
    
    params = isotopes[name]
    energies = params["energies"]
    intensities = params["intensities"]
    activity = params["activity"]
    time_lst = params["time"]
    time = time_lst[0] if detector == "CeBrLa" else time_lst[1]

    peak_index = peak_counter[detector].get(name)
    E_gamma = energies[peak_index]
    I_gamma = intensities[peak_index]

    initial_guess = [max(y), x[np.argmax(y)], 5, 0, min(y)]
    popt, pcov = curve_fit(model, x, y, sigma = y_err, p0=initial_guess, absolute_sigma=True)
    A, mu, sigma, B, C = popt
    A_err, mu_err, sigma_err, B_err, C_err = np.sqrt(np.diag(pcov))
    
    K = activity * I_gamma / 100 * time
    eff = A * 100 / K
    eff_err = A_err * 100 / K
    
    result_entry = [(E_gamma, I_gamma, activity), (mu, mu_err), (sigma, sigma_err), (eff, eff_err)]
    results_dict = CeBrLa_results if detector == "CeBrLa" else CeBr_results
    if name not in results_dict:
        results_dict[name] = []
    results_dict[name].append(result_entry)

    chi_sq = np.sum(((y - model(x, *popt)) / y_err)**2)
    dof = len(x) - len(popt)
    reduced_chi_sq = chi_sq / dof
    p_value = 1 - chi2.cdf(chi_sq, dof)

    plt.figure(figsize=(6,3))
    plt.errorbar(x, y, yerr=y_err, fmt='o', label='Data', capsize=4, color='black', zorder=1)
    plt.plot(x, model(x, *popt), 'r--', linewidth=2, label="Gaussian + linear fit", zorder=3)

    plt.xlabel("Channel")
    plt.ylabel("Counts")
    plt.title(f"{name} peak {peak_index + 1}, {detector}")
    plt.legend()
    plt.show()

    print(f"Counts = {A:.5g} ± {A_err:.5g}")
    print(f"Mean = {mu:.5g} ± {mu_err:.5g}")
    print(f"Sigma = {sigma:.5g} ± {sigma_err:.5g}")
    print(f"Background linear = {B:.5g} ± {B_err:.5g}")
    print(f"Background constant = {C:.5g} ± {C_err:.5g}")
    print("---------------------------------")
    print("Reduced chi squared = ", reduced_chi_sq)
    print("P-value = ", p_value)
```

```python
def sort_peaks(results_dict):
    combined = []
    colors = {
    "Co60": "blue",
    "Cs137": "magenta",
    "Na22": "purple",
    "Eu152": "orange"
    }
    
    for isotope, entries in results_dict.items():
        for entry in entries:
            energy, intesity, activity = entry[0]
            mu, mu_err = entry[1]
            sigma, sigma_err = entry[2]
            eff, eff_err = entry[3]
            color = colors[isotope]

            combined.append({
                "energy" : energy,
                "intesity" : intesity,
                "activity" : activity,
                "mu": mu,
                "mu_err": mu_err,
                "sigma": sigma,
                "sigma_err": sigma_err,
                "eff": eff,
                "eff_err": eff_err,
                "isotope": isotope,
                "color": color
            })

    combined_sorted = sorted(combined, key=lambda x: x["mu"])

    return combined_sorted
```

```python
def calc_E_res(mu, mu_err, sigma, sigma_err, b, c, b_err, c_err):
    E = b*mu + c
    E_err = np.sqrt((mu * b_err)**2 + (b * mu_err)**2 + (c_err)**2)
    
    FWHM = 2.355 * (b*sigma + c)
    FWHM_err = 2.355 * np.sqrt((sigma * b_err)**2 + (b * sigma_err)**2 + (c_err)**2)
    
    E_res = FWHM/E * 100
    E_res_err = np.sqrt((FWHM_err / E)**2 + (FWHM * E_err / E**2)**2) * 100
    
    return E_res, E_res_err
```

```python
E_res_CeBrLa, E_res_err_CeBrLa = calc_E_res(mu_CeBrLa, mu_err_CeBrLa, sigma_CeBrLa, sigma_err_CeBrLa, 
                                            B_CeBrLa, C_CeBrLa, B_err_CeBrLa, C_err_CeBrLa)

E_res_CeBr, E_res_err_CeBr = calc_E_res(mu_CeBr, mu_err_CeBr, sigma_CeBr, sigma_err_CeBr, 
                                            B_CeBr, C_CeBr, B_err_CeBr, C_err_CeBr)
```

```python
def resolution_fit(x, M, B):
    return M/np.sqrt(x) + B
```

```python
y = E_res_CeBrLa
y_err = E_res_err_CeBrLa

initial_guess = [1.2, 0.8]
popt, pcov = curve_fit(resolution_fit, x, y, sigma=y_err, p0=initial_guess, absolute_sigma=True)
M, B = popt
M_err, B_err = np.sqrt(np.diag(pcov))

fit_vals = resolution_fit(x, M, B)
residuals = y - fit_vals

chi_sq = np.sum(((y - fit_vals) / y_err)**2)
dof = len(x) - len(popt)
reduced_chi_sq = chi_sq / dof
p_value = 1 - chi2.cdf(chi_sq, dof)

x_smooth = np.linspace(min(x), max(x), 500)
fit_smooth = resolution_fit(x_smooth, M, B)

fig, (ax1, ax2) = plt.subplots(
    2, 1, figsize=(8, 6), sharex=True,
    gridspec_kw={"height_ratios": [3, 1]}  
)


used_labels = set()
for i, p in enumerate(CeBrLa_fits):
    label = p["isotope"] if p["isotope"] not in used_labels else None
    used_labels.add(p["isotope"])
    ax1.errorbar(x[i], y[i], yerr=y_err[i], fmt='o', color=p["color"], capsize=4, label=label)
    
ax1.plot(x_smooth, fit_smooth, 'r--', linewidth=2, label="Linear fit", zorder=3)
ax1.set_ylabel("Resolution (%)")
ax1.set_title("CeBrLa Resolution")
ax1.legend()
ax1.grid(True, alpha=0.3)

used_labels = set()
for i, p in enumerate(CeBrLa_fits):
    label = p["isotope"] if p["isotope"] not in used_labels else None
    used_labels.add(p["isotope"])
    ax2.errorbar(x[i],  residuals[i], yerr=y_err[i], fmt='o', color=p["color"], capsize=4, label=label)
    
ax2.axhline(0, color='r', linewidth=1, zorder=2)
ax2.set_xlabel("Energy [keV]")
ax2.set_ylabel("Residual Resolution (%)")
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"M = {M:.5g} ± {M_err:.5g}")
print(f"B = {B:.5g} ± {B_err:.5g}")
print("----------------------")
print("Reduced chi squared = ", reduced_chi_sq)
print("P-value = ", p_value)
```
